## c) Exploratory Data Analysis

## Visitor Preferences:
### 1) What types of establishments (historical sites, natural attractions, modern landmarks, etc.) are more popular among visitors?

```python
import matplotlib.pyplot as plt

# Get the value counts and sort them by count in descending order
significance_counts = data['Significance'].value_counts().sort_values(ascending=True)

# Plot the horizontal bar chart
significance_counts.plot.barh()
plt.title('Significance Distribution')
plt.xlabel('Count')
plt.ylabel('Significance')
plt.show()
```

## Observation: 

The Top 5 most popular visiting establishments are Historical, Religious, Nature, Recreational and Wildlife

## 2) Are there correlations between entrance fee, time needed to visit, and Google review ratings?

```python
corr_matrix=data.corr(numeric_only=True)
corr_matrix
```

```python
corr_matrix["Time needed to visit in hrs"]
```

```python
import seaborn as sns
sns.heatmap(data.corr(numeric_only=True).round(2),annot=True,vmin=-1,vmax=1,cmap='vlag')
```

## Observation:
a) "Time needed to visit in hrs" is positively correlated with the "Entrance fee in INR". This shows that if time need to visit is more then the entrance fees is also more.  
b) "Time needed to visit in hrs" is negatively correlated with the "Google review rating". This shows that if time need to visit is more Google review rating is less.

## 3) What is the average time visitors spend at these locations? How does this vary across different types of establishments?

```python
data.groupby('Significance')['Time needed to visit in hrs'].mean().sort_values(ascending= False)
```

```python

# Group by 'Significance' and calculate the mean time needed to visit
mean_time_by_significance = data.groupby('Significance')['Time needed to visit in hrs'].mean()

# Sort the values in order, plot the top 10, and then display the plot
mean_time_by_significance.sort_values(ascending= True).tail(10).plot.barh(color='magenta')

plt.title('Top 10 Significance Categories by Mean Time Needed to Visit')
plt.xlabel('Mean Time Needed to Visit (hrs)')
plt.ylabel('Significance')
plt.show()
```

```python
data.groupby('Type')['Time needed to visit in hrs'].mean().sort_values(ascending= False)
```

```python
#Group by 'Type' and calculate the mean time needed to visit
mean_time_by_type = data.groupby('Type')['Time needed to visit in hrs'].mean()

# Sort the values in order, plot the top 10, and then display the plot
mean_time_by_type.sort_values(ascending= True).tail(10).plot.barh()

plt.title('Top 10 Type Categories by Mean Time Needed to Visit')
plt.xlabel('Mean Time Needed to Visit (hrs)')
plt.ylabel('Type')
plt.show()
```

## Observation:
More amount of time spent in Education, Entertainment, Wildlife, Trekking, Sports

```python
data.groupby('Type')['Time needed to visit in hrs'].mean().sort_values(ascending= False)
```

```python

# Group by 'Significance' and calculate the mean time needed to visit
mean_time_by_significance = data.groupby('Significance')['Time needed to visit in hrs'].mean()

# Sort the values in order, plot the top 10, and then display the plot
mean_time_by_significance.sort_values(ascending= False).tail(10).plot.barh(color='red')

plt.title('Top 10 Significance Categories by  Least Mean Time Needed to Visit')
plt.xlabel('Mean Time Needed to Visit (hrs)')
plt.ylabel('Significance')
plt.show()
```

## Observation:
a) Least amount of time spent on Architectural, Religious (temple), Spiritual, Engineering marvel, Botanical , Nature related.

## Plot the histogram of Google review rating

```python

plt.figure(figsize=(10, 6))
plt.hist(data['Google review rating'], bins=20, color='orange', edgecolor='black')

plt.title('Distribution of Google Review Ratings')
plt.xlabel('Google Review Rating')
plt.ylabel('Frequency')

plt.show()
```

## Observation:
a) There found to be one outlier at google review rating of 1.5  
b) Apart from that outlier, the distribution is left skewed. This skewness pattern indicates that there are relatively more extreme values on the left side of the distribution, and the bulk of the values are on the right.  
c) The mean (average) is typically less than the median, and the median may be closer to the right tail.
